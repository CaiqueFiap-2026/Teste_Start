##Funções de produto

```python
class Produto:
  def __init__(self, id, nome, marca, tipo, quantidade, preco):
    self.id = id
    self.nome = nome
    self.marca = marca
    self.tipo = tipo
    self.quantidade = quantidade
    self.preco = preco
    self.desconto = 0

  def __str__(self):
    return (f"[{self.id}] {self.nome} | "
                f"Marca: {self.marca} | "
                f"Tipo: {self.tipo} | "
                f"Estoque: {self.quantidade} | "
                f"Preço: R$ {self.preco:.2f} | "
                f"Desconto: {self.desconto}%")
```

```python
def menu_produtos():
    global produtos, contador
    while True:
        print("\n===== Produtos =====")
        print("1 - Cadastrar produto")
        print("2 - Listar produtos")
        print("3 - Buscar produto")
        print("4 - Atualizar produto")
        print("5 - Remover produto")
        print("6 - Adicionar quantidade")
        print("7 - Reduzir quantidade")
        print("8 - Exportar lista de produtos")
        print("9 - Importar lista de produtos")
        print("10 - Aplicar desconto")
        print("11 - Remover desconto")
        print("0 - Voltar")

        opcao = input("\nEscolha uma opção: ")

        match opcao:
            case "1":
                produtos, contador = cadastrar_produto(produtos, contador)
            case "2":
                listar_produtos(produtos)
            case "3":
                id = int(input("ID do produto: "))
                produto = buscar_produto(produtos, id)
                print(produto)
            case "4":
                id = int(input("ID do produto: "))
                produtos = atualizar_produto(produtos, id)
            case "5":
                id = int(input("ID do produto: "))
                produtos = remover_produto(produtos, id)
            case "6":
                id = int(input("ID do produto: "))
                produtos = adicionar_quantidade(produtos, id)
            case "7":
                id = int(input("ID do produto: "))
                produtos = reduzir_quantidade(produtos, id)
            case "8":
                exportar_produtos(produtos)
            case "9":
                produtos = importar_produtos()
            case "10":
                id = int(input("ID do produto: "))
                percentual = float(input("Percentual de desconto: "))
                produtos = aplicar_desconto(produtos, id, percentual)
            case "11":
                id = int(input("ID do produto: "))
                produtos = remover_desconto(produtos, id)
            case "0":
                print("Retornando ao menu principal...")
                break
            case _:
                print("Opção inválida. Tente novamente.")
```

```python
def cadastrar_produto(produtos: list, contador_produtos: int) ->  tuple:
  """
  Cadastra um novo produto no sistema

  Args:
    produtos (list): Lista de produtos já cadastrados no sistema.
    contador_produtos (int): Contador para geração automática de ID dos produtos.

  Retorns:
    tuple: A lista com os valores atualizados de produtos e contador de ID.
  """
  nome = input("Nome do produto: ")
  marca = input("Marca: ")
  tipo = input("Tipo: ")

  try:
    quantidade = int(input("Quantidade em estoque: "))
    preco = float(input("Preço: R$ "))
  except ValueError:
    print("Erro: digite um valor válido!")
    return produtos, contador_produtos

  produto = Produto(contador_produtos, nome, marca, tipo, quantidade, preco)
  produtos.append(produto)
  print(f"Produto '{nome}' cadastrado com sucesso!")
  return produtos, contador_produtos + 1
```

```python
def listar_produtos(produtos):
    if len(produtos) == 0:
        print("Nenhum produto cadastrado.")
        return
    for produto in produtos:
        print(produto)
```

```python
def buscar_produto(produtos, id):
    for produto in produtos:
        print(produto.id)
        if produto.id == id:
            return produto
    return None
```

```python
def atualizar_produto(produtos, id):
  produto = buscar_produto(produtos, id)
  if produto == None:
    print("Produto não encontrado!")
    return produtos

  print(f"Atualizando: {produto}")
  nome = input(f"Novo nome ({produto.nome}): ")
  marca = input(f"Nova marca ({produto.marca}): ")
  tipo = input(f"Novo tipo ({produto.tipo}): ")
  preco = input(f"Novo preço ({produto.preco}): ")

  produto.nome = nome
  produto.marca = marca
  produto.tipo = tipo
  produto.preco = float(preco)

  print("Atualizado com sucesso!")
  return produtos
```

```python
def remover_produto(produtos, id):
    produto = buscar_produto(produtos, id)
    if produto is None:
        print("Produto não encontrado.")
        return produtos
    produtos.remove(produto)
    print(f"Produto '{produto.nome}' removido com sucesso!")
    return produtos
```

```python
def adicionar_quantidade(produtos, id):
    produto = buscar_produto(produtos, id)
    if produto is None:
        print("Produto não encontrado.")
        return produtos
    qtd = int(input("Quantidade a adicionar: "))
    produto.quantidade += qtd
    print(f"Estoque atualizado. Nova quantidade: {produto.quantidade}")
    return produtos
```

```python
def reduzir_quantidade(produtos, id):
    produto = buscar_produto(produtos, id)
    if produto is None:
        print("Produto não encontrado.")
        return produtos
    qtd = int(input("Quantidade a reduzir: "))
    if qtd > produto.quantidade:
      print("Opa, quantidade insuficiente!")
      return produtos
    print(f"Estoque atualizado. Nova quantidade: {produto.quantidade}")
    return produtos
```

```python
def exportar_produtos(produtos):
  dados = []
  for produto in produtos:
    dici = {
        "id": produto.id,
        "nome": produto.nome,
        "marca": produto.marca,
        "tipo": produto.tipo,
        "quantidade_estoque": produto.quantidade,
        "preco": produto.preco,
        "desconto": produto.desconto
    }
    dados.append(dici)

  with open("produtos.json", "w", encoding="utf-8") as arquivo:
    json.dump(dados, arquivo, indent=4, ensure_ascii=False)
  print("Produtos exportados com sucesso!")
```

```python
def importar_produtos():
    try:
        with open("produtos.json", "r", encoding="utf-8") as arquivo:
            dados = json.load(arquivo)
        produtos = []
        for item in dados:
            produto = Produto(
                item["id"], item["nome"], item["marca"],
                item["tipo"], item["quantidade_estoque"], item["preco"]
            )
            produto.desconto = item["desconto"]
            produtos.append(produto)
        return produtos
    except:
        return []
```

```python
def aplicar_desconto(produtos, id, percentual):
  produto = buscar_produto(produtos, id)
  if produto is None:
      print("Produto não encontrado.")
      return produtos

  if percentual <= 0 or percentual > 100:
      print("Erro: informe um percentual entre 1% e 100%.")
      return produtos

  produto.desconto = percentual
  print(f"O novo desconto do produto {produto.nome} é {produto.desconto}")
  return produtos
```

```python
def remover_desconto(produtos, id):
  produto = buscar_produto(produtos, id)
  if produto is None:
      print("Produto não encontrado.")
      return produtos

  produto.desconto = 0
  print(f"O desconto do produto {produto.nome} foi removido")
  return produtos
```

##Funções de cliente

```python
def menu_clientes():
    global clientes, contador_cliente

    while True:
        print("\n===== Clientes =====")
        print("1 - Cadastrar cliente")
        print("2 - Listar clientes")
        print("3 - Buscar cliente")
        print("4 - Exportar lista de clientes")
        print("5 - Importar lista de clientes")
        print("0 - Voltar")

        opcao = input("\nEscolha uma opção: ")

        match opcao:
            case "1":
                clientes, contador_cliente = cadastrar_cliente(clientes, contador_cliente)
            case "2":
                listar_clientes(clientes)
            case "3":
                id = int(input("ID do cliente: "))
                cliente = buscar_cliente(clientes, id)
                if cliente:
                    print(cliente)
                else:
                    print("Cliente não encontrado.")
            case "4":
                exportar_clientes(clientes)
            case "5":
                clientes = importar_clientes()
            case "0":
                print("voltando...")
                break
            case _:
                print("Opção inválida. Tente novamente.")
```

```python
class Cliente:
    def __init__(self, id, nome, cpf, telefone):
        self.id = id
        self.nome = nome
        self.cpf = cpf
        self.telefone = telefone

    def __str__(self):
        return (f"[{self.id}] {self.nome} | "
                f"CPF: {self.cpf} | "
                f"Telefone: {self.telefone}")
```

```python
def cadastrar_cliente(clientes, contador_id):
    nome = input("Nome do cliente: ")
    cpf = input("CPF: ")
    telefone = input("Telefone: ")
    cliente = Cliente(contador_id, nome, cpf, telefone)
    clientes.append(cliente)
    print(f"Cliente '{nome}' cadastrado com sucesso!")
    return clientes, contador_id + 1
```

```python
def listar_clientes(clientes):
    if len(clientes) == 0:
        print("Nenhum cliente cadastrado.")
        return
    for cliente in clientes:
        print(cliente)
```

```python
def buscar_cliente(clientes, id):
    for cliente in clientes:
        if cliente.id == id:
            return cliente
    return None
```

```python
def exportar_clientes(clientes):
    dados = []
    for cliente in clientes:
        dados.append({
            "id": cliente.id,
            "nome": cliente.nome,
            "cpf": cliente.cpf,
            "telefone": cliente.telefone
        })
    with open("clientes.json", "w", encoding="utf-8") as arquivo:
        json.dump(dados, arquivo, indent=4, ensure_ascii=False)
    print("Clientes exportados com sucesso!")
```

```python
def importar_clientes():
    try:
        with open("clientes.json", "r", encoding="utf-8") as arquivo:
            dados = json.load(arquivo)
        clientes = []
        for item in dados:
            cliente = Cliente(
                item["id"], item["nome"],
                item["cpf"], item["telefone"]
            )
            clientes.append(cliente)
        return clientes
    except:
        return []
```

##Funções de venda

```python
def menu_vendas():
    global produtos, clientes, historico

    while True:
        print("\n===== Vendas =====")
        print("1 - Registrar venda")
        print("2 - Listar histórico")
        print("0 - Voltar")

        opcao = input("\nEscolha uma opção: ")

        match opcao:
          case "1":
              produtos, historico = registrar_venda(produtos, clientes, historico)
          case "2":
              listar_vendas(historico)
          case "0":
              print("Retornando ao menu principal...")
              break
          case _:
            print("deu errado!")
```

```python
def registrar_venda(produtos, clientes, historico):
  id_cliente = int(input("ID do cliente: "))
  cliente = buscar_cliente(clientes, id_cliente)
  if cliente is None:
      print("Cliente não encontrado.")
      return produtos, historico

  itens = []
  valor_total = 0

  while True:
    id_produto = input("ID do produto (ou 0 para finalizar): ")
    if id_produto == "0":
      break

    produto = buscar_produto(produtos, int(id_produto))
    if produto is None:
      print("Produto não encontrado.")
      continue

    qtd = int(input(f"Quantidade de '{produto.nome}': "))
    if qtd > produto.quantidade:
      print(f"Estoque insuficiente. Disponível: {produto.quantidade}")
      continue

    produto.quantidade -= qtd
    itens.append((produto.nome, qtd))
    valor_total += produto.preco * (1-produto.desconto/100) * qtd

  if len(itens) == 0:
    print("Nenhum item adicionado. Venda cancelada.")
    return produtos, historico

  venda = {
      "produtos": itens,
      "valor_total": valor_total
  }

  historico.append(venda)
  return produtos, historico
```

```python
def listar_vendas(historico):
  if len(historico) > 0:
    cont_vendas = 0
    for venda in historico:
      cont_vendas += 1
      print(f"{cont_vendas} | Produtos: {venda["produtos"]} | Valor total R$ {venda["valor_total"]:.2f}")

  else:
    print("Nenhuma venda registada nesta sessão. Retornando...")
```

```python
def fechamento(historico):
  if len(historico) == 0:
    print("Nenhuma venda registrada nesta sessão.")
    return

  total = 0
  #venda é produtos, valor_total
  for venda in historico:
    total += venda["valor_total"]

  resumo = {
      "data_hora": datetime.now().strftime("%d/%m/%Y %H:%M:%S"),
      "total_vendas": len(historico),
      "valor_total_recebido": total,
      "vendas": historico
  }

  print(f"Resumo de hoje {resumo["data_hora"]}:")
  print(f"Total de vendas: {resumo["total_vendas"]}")
  print(f"Valor total recebido: {resumo["valor_total_recebido"]}")
  print(f"Estas foram as vendas de hoje:")
  listar_vendas(historico)

  nome_arquivo = f"fechamento_{datetime.now().strftime("%d-%m-%Y_%H-%M-%S")}.json"
  with open(nome_arquivo, "w", encoding="utf-8") as arquivo:
    json.dump(resumo, arquivo, indent=4, ensure_ascii=False)
  print(f"Fechamento exportando com sucesso para {nome_arquivo}.")
```
